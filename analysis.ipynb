```
%sql
select 
  distinct * 
from methods
```

```
sparkdf = spark.table("methods")
sparkdf = sparkdf.distinct()
df = sparkdf.toPandas()
```

```
import seaborn as sns
from matplotlib import pyplot as plt
sns.set(style="white")
sns.set_style("whitegrid")

#define dimensions of subplots (rows, columns)
fig, axes = plt.subplots(1, 2, figsize=(15,5),sharey=True)

sns.barplot(ax=axes[0],
            x='method',
            y='duration',
            hue='number_disks',
            order=['sequential','concurrent.futures','pandas UDFs'],
            data=df[
                (df['mode']=='single node') & 
                (df['algorithm']=='linear regression')
            ])
axes[0].set_title('Linear regression, 1 node')

sns.barplot(ax=axes[1],
            x='method',
            y='duration',
            hue='number_disks',
            order=['sequential','concurrent.futures','pandas UDFs'],
            data=df[
                (df['mode']=='four workers') & 
                (df['algorithm']=='linear regression')
            ])
axes[1].set_title('Linear regression, 4 workers')


fig2, axes2 = plt.subplots(1, 2, figsize=(15,5),sharey=True)
sns.barplot(ax=axes2[0],
            x='method',
            y='duration',
            hue='number_disks',
            order=['sequential','concurrent.futures','pandas UDFs'],
            data=df[
                (df['mode']=='single node') & 
                (df['algorithm']=='fbprophet')
            ])
axes2[0].set_title('fbprophet, 1 node')

sns.barplot(ax=axes2[1],
            x='method',
            y='duration',
            hue='number_disks',
            order=['sequential','concurrent.futures','pandas UDFs'],
            data=df[
                (df['mode']=='four workers') & 
                (df['algorithm']=='fbprophet')
            ])
axes2[1].set_title('fbprophet, 4 workers')

fig3, axes3 = plt.subplots(1, 2, figsize=(15,5),sharey=True)
sns.barplot(ax=axes3[0],
            x='method',
            y='duration',
            hue='number_disks',
            order=['sequential','concurrent.futures','pandas UDFs'],
            data=df[
                (df['mode']=='single node') & 
                (df['algorithm']=='combined')
            ])
axes3[0].set_title('combined, 1 node')

sns.barplot(ax=axes3[1],
            x='method',
            y='duration',
            hue='number_disks',
            order=['sequential','concurrent.futures','pandas UDFs'],
            data=df[
                (df['mode']=='four workers') & 
                (df['algorithm']=='combined')
            ])
axes3[1].set_title('combined, 4 workers')
```
